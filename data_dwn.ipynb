```python
import os
import yfinance as yf
import pandas as pd
```

```python
df=pd.read_excel("names.xlsx")
```

```python
df.dropna(inplace=True)
```

```python
names=df['Symbol'].to_list()
```

```python
for count in range(len(names)):
    names[count] = names[count] + ".NS"
```

```python
nifty_tickers = yf.Tickers(names).tickers
```

```python
top_nifty_tickers = [nifty_tickers[t].ticker for t in nifty_tickers]
```

```python
daily_data = yf.download(top_nifty_tickers[0], period="max", interval="1d")
```

```python
daily_data
```

```python
for i in top_nifty_tickers:
    daily_data = yf.download(i, period="max", interval="1d")
    weekly_data = yf.download(top_nifty_tickers, period="max", interval="1wk")
    daily_data.to_csv("data/daily/"+i+".csv")
    weekly_data.to_csv("data/weekly/"+i+".csv")
```

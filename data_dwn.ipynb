```python
import os
import yfinance as yf
import pandas as pd
import pywt
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
df=pd.read_excel("names.xlsx")
```

```python
df.dropna(inplace=True)
```

```python
names=df['Symbol'].to_list()
```

```python
for count in range(len(names)):
    names[count] = names[count] + ".NS"
```

```python
nifty_tickers = yf.Tickers(names).tickers
```

```python
top_nifty_tickers = [nifty_tickers[t].ticker for t in nifty_tickers]
```

```python
daily_data = yf.download(top_nifty_tickers[0], period="max", interval="1d")
```

```python
daily_data
```

```python
for i in range(len(top_nifty_tickers)):
    daily_data = yf.download(top_nifty_tickers[i], period="max", interval="1d")
    weekly_data = yf.download(top_nifty_tickers[i], period="max", interval="1wk")
    daily_data.to_csv("data/daily/"+top_nifty_tickers[i]+".csv")
    weekly_data.to_csv("data/weekly/"+top_nifty_tickers[i]+".csv")
```

```python
names=[]
for root, dirs, files in os.walk(".\data\daily"):
    names=files
```

```python
os.mkdir('./transformed_data')
```

```python
df=pd.read_csv(f'./data/daily/{names[0]}')
df.Close
```

```python
os.mkdir('./transformed_data/daily')
omitted=[]
for i in range(len(names)):
    df=pd.read_csv(f'./data/daily/{names[i]}')
    df.dropna()
    if len(df.Close)>0:
        (ca,cd)=pywt.dwt(df.Close,'haar')
        ca=np.nan_to_num(ca)
        cd=np.nan_to_num(cd)
        plt.imsave(f'./transformed_data/daily/{names[i]}.png',(ca,cd))
    else:
        omitted.append(names[i])
```

```python
omitted=[]
for i in range(len(names)):
    df=pd.read_csv(f'./data/daily/{names[i]}')
    df.dropna()
    if len(df.Close)<=0:
        omitted.append(names[i])    
```

```python
os.mkdir('./transformed_data/weekly')
omitted_again=[]
for i in range(len(names)):
    df=pd.read_csv(f'./data/weekly/{names[i]}')
    df.dropna()
    if len(df.Close)>0 and names[i] not in omitted:
        (ca,cd)=pywt.dwt(df.Close,'haar')
        ca=np.nan_to_num(ca)
        cd=np.nan_to_num(cd)
        plt.imsave(f'./transformed_data/weekly/{names[i]}.png',(ca,cd))
    else:
        omitted_again.append(names[i])
```

```python
os.mkdir('./lr')
dict={}
for i in range(len(names)):
     dict[names[i]]=i
     if names[i] not in omitted:
          im=plt.imread(f'./transformed_data/daily/{names[i]}.png')
          k=cv2.resize(im,(32,32))
          plt.imsave(f'./lr/{i}.png',k)
```

```python
os.mkdir('./hr')
for i in range(len(names)):
     if names[i] not in omitted:
          im=plt.imread(f'./transformed_data/weekly/{names[i]}.png')
          k=cv2.resize(im,(128,128))
          plt.imsave(f'./hr/{i}.png',k)
```
